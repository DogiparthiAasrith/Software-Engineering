```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization,Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.inception_v3 import InceptionV3
import tensorflow as tf
```

```python
# Define the image dimensions and batch size
image_size = (224,224)
batch_size = 64
```

```python
# Data Augmentation

train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

valid_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
train_data = train_datagen.flow_from_directory(
    directory='/kaggle/input/potatoclass/Dataset/Dataset/train',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
)

validation_data = valid_datagen.flow_from_directory(
    directory='/kaggle/input/potatoclass/Dataset/Dataset/val',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
  )
```

```python
inception_model = InceptionV3(input_shape=(224, 224, 3),
                        include_top=False,
                        weights='imagenet')

for layer in inception_model.layers:
    layer.trainable = False

# Add custom layers
inception_x= inception_model.output

x = Flatten()(inception_x)
x=BatchNormalization()(x)
x= Dense(3, activation='softmax')(x)  # 4 categories

# Create new model
inception_model = Model(inputs=inception_model.input, outputs=x)
```

```python
# Compile the model
inception_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    metrics=['accuracy'],
    loss='categorical_crossentropy'
)
```

```python
# train the model
inception_model.fit(train_data,
          validation_data=validation_data,
          epochs=30,
          batch_size=100,
          steps_per_epoch=len(train_data),
          validation_steps=len(validation_data),
          #callbacks=[model_checkpoint]
        )
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
y_pred_prob = inception_model.predict(validation_data)
y_pred = np.argmax(y_pred_prob, axis=1)
y_true = validation_data.classes

# Calculate performance metrics
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')
accuracy = accuracy_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)

print("\nInceptionV3 Performance Metrics:")
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)
```
